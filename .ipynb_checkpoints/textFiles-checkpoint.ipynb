## Text Files in Python

A file is a collection of data that is stored on disk

Data can be in text format or in binary format

Binaries can contain: compiled program code, images, video, compressed files

Text files are humanly readable but can contain control characters like newline (\n) or tabs (\n) 

Text files can all be opened and read in text editors (like notepad), Binary files need specific applications that can open and interpret the binary data in a human readable format


```python
#opening a file - it must be in the same directory (folder)
file = open("test1.txt", "r") # notice: this text file has been created for you - open it and see it, add to it
```

```python
#read the entire contents into a string
contents= file.read()
print(contents)
file.close()
```

```python
file = open("test1.txt", "r")

#read the entire contents into a List of  strings
linesList=file.readlines()
print(linesList)

#close the file
file.close()
```

```python
# now do something with the list
for i in range(len(linesList)):
    print(" line number ",i, "=", linesList[i])
```

```python
# now do something with each character of each line (string)
for i in range(len(linesList)):
    for ch in linesList[i]:
        print(ch.lower(),end="*")
    print() # go to a new line after each line in the file
```

# Open the file and print the length of the each word on a new line


```python
file = open("test1.txt", "r")
contents= file.readlines()
for i in range(len(contents)):
    print(len(contents[i].replace('\n', '')))

file.close()
```

# Open the file and print the word backwards


```python
file = open("test1.txt", "r")
line = file.read()
print(line[::-1])

file.close()
```

# Open the file and print the vowels of each word, use stars (*) for the consonants (eg banana will be printed as *a*a*a )

```python
file = open("test1.txt", "r")
Llist = file.readlines()

for i in range(len(Llist)):
    Llist[i] = Llist[i].replace('\n', '')

vowels = ['a', 'e', 'i', 'o', 'u']
for i in range(len(Llist)):
    for ch in Llist[i]:
        if ch in vowels:
            print(ch, end='')
        else:
            print('*',end='')
    print()
file.close()
```
